#### now we will do lemmatization ..changing the each word to its root word...using part of speech of each word.

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk import pos_tag
```

```python
# import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
w = "better"
pos_tag([w])
```

```python

# nltk.download('wordnet')
from nltk.corpus import wordnet

def get_simple_pos(tag):
    
    if tag.startswith('J'):
        return wordnet.ADJ
    if tag.startswith('V'):
        return wordnet.VERB
    if tag.startswith('N'):
        return wordnet.NOUN
    if tag.startswith('R'):
        return wordnet.ADV
    
    return wordnet.NOUN
    
```

#### getting all stopwords in english language and all pucnhuations to remove it from the array of words(review).

```python

# nltk.download('stopwords')
from nltk.corpus import stopwords
import string
stops =set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
stops
```

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w ,pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())   
    return output_words
    
    
```

```python
documents = [(clean_review(document), category) for document, category in documents]
```

```python
## Now we cleaned the reviews..it is ready to use the data for training it on models.
```

```python
## first we have to vectorize our datasets..that we first have to convert in tha type so that we can apply iton sklearn models.
```

### we will use the dfidf vectorizer for vectorization.

```python
#tf-Idf vectorizer

from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features = 1000) 
```

```python
categories = [category for document, category in documents]
```

```python
text_documents = [" ".join(document) for document, category in documents]
```

```python
from sklearn import model_selection
```

```python
x_train ,x_test ,y_train ,y_test = model_selection.train_test_split(text_documents , categories)
```

```python
x_train_features = vectorizer.fit_transform(x_train)
x_train_features.todense()
```

```python
vectorizer.get_feature_names()
```

```python
x_test_features = vectorizer.transform(x_test)
```

```python
x_test_features
```

### implementing svm sklearn model now.

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
svc.fit(x_train_features , y_train)
```

```python
y_pred = svc.predict(x_test_features)
```

```python
svc.score(x_test_features ,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test ,y_pred)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test , y_pred))
```
